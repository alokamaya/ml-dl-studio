## 📂 Data Loading - Breast Cancer Dataset

In this step, we load the Breast Cancer dataset provided by scikit-learn.  
This dataset is commonly used for binary classification tasks.

### 📊 Dataset Overview
- The dataset contains features computed from digitized images of breast masses.
- Each sample is labeled as:
  - **0 → Malignant (cancerous)**
  - **1 → Benign (non-cancerous)**

### 🧠 What we are doing
- Load the dataset using `load_breast_cancer()`
- Convert it into a pandas DataFrame for easier analysis
- Add the target column to the DataFrame

### 🎯 Objective
Prepare the dataset for further analysis and model building.

```python
from sklearn.datasets import load_breast_cancer
import pandas as pd

data = load_breast_cancer()
#data
df = pd.DataFrame(data.data, columns=data.feature_names)
df['target'] = data.target

df.tail()
```
